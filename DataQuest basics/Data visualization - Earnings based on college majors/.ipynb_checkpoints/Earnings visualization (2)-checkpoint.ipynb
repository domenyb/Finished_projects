This guided project was a visualization project of the Data Scientist path.
The data is job outcomes of students who graduated between 2010 and 2012, and the original (uncleaned) dataset was released as an American Community Survey.

The questions I shall explore are mainly: 
    - Do students of the more popular majors make more? 
    - Which majors have significant disparities in gender?
    - Which majors are the most popular overall?

From the scatter-plots at first I shall try to clear a few relations before moving forward to look at the distributions in more detail.

The median income is not the highest for the most popular majors - though this effect can be explained if there is an overabundance of those majors already...

On the other hand, the unemployment rate is also relatively low in those areas, and some others have a much higher unemployment rate. 

There seems to be an anticorrelation between the amount of women in the majors and the median salary.

A clear link between the amount of full-time workers and the median salary currently can not be estabilished, one would need to dive deeper for that.


As the next step, I shall look at the distributions of the above used metrics in the form of histograms. 

Now I will do this in different cells since the data-ranges (and probably the bins as well) would require special attention.... but the subplots still provide a more compact way to look at it imho...

```python
metric="Sample_size"
recent_grads[metric].hist(range=[0,3000], bins=20)# I have excluded a single 4000+ major
plt.title("Distribution of "+ metric.replace("_", " "))
```

```python
metric="Median"
recent_grads[metric].hist(range=[20000,85000], bins=20)#Once again, I have excluded a single outlier
plt.title("Distribution of "+ metric.replace("_", " "))
```

```python
metric="Employed"
recent_grads[metric].hist(range=[0,200000], bins=20)# Once again a single exclusion
# I could probably experiment for even better resolution at the bottom if I suspected this to be a
#very important metric.
plt.title("Distribution of "+ metric.replace("_", " "))
```

```python
metric="Full_time"
recent_grads[metric].hist(range=[0,18000], bins=20)# One exclusion for a much better distribution
plt.title("Distribution of "+ metric.replace("_", " "))
```

```python
metric="ShareWomen"
recent_grads[metric].hist(bins=10)# This is fine, there isn't much more info
# in using 15-20 bins, the shape of the distribution stays similar.
plt.title("Distribution of "+ metric.replace("_", " "))
```

```python
metric="Unemployment_rate"
recent_grads[metric].hist(bins=30)# This is an important metric, I think 30 bins
#could represent it just fine. I have tried 50, but that just introduces "noise"
plt.title("Distribution of "+ metric.replace("_", " "))
```

```python
metric="Men"
recent_grads[metric].hist(range=[0,10000], bins=20)# I have excluded some, but this way I have a much better 
#look at the bulk of the sample
plt.title("Distribution of "+ metric.replace("_", " "))
```

```python
metric="Women"
recent_grads[metric].hist(range=[0,20000], bins=20)# Interestingly, here I have to make
# the cutoff at twice the same value - but once again, this is a nice insight at the distributions.
plt.title("Distribution of "+ metric.replace("_", " "))
```

I think at first glance the most interesting thing to look at is the difference in distribution of men and women. 

Men don't really seem to be represented at as high numbers as women, but at the same time women seem to be virtually "absent" from a much higher number of fields than men.
